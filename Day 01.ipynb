# Advent of Code 2023, Day 1
[Day 1 Challenge](https://adventofcode.com/2023/day/1)

```python
import aoc
%reload_ext autoreload
```

```python
input_list = aoc.split_contents(aoc.read_input('Input/day_01.txt'))
```

```python
def get_first_digit(input_string: str) -> int:
    for i, c in enumerate(input_string):
        if c.isdigit():
            break
    return int(c)
```

# Part 1

```python
sum_of_calibration_values = 0
for item in input_list:
    first_digit = get_first_digit(item)
    last_digit = get_first_digit(item[::-1])
    calibration_value = first_digit * 10 + last_digit
    print(f'{calibration_value}')
    sum_of_calibration_values += calibration_value
```

```python
sum_of_calibration_values
```

# Part 2

```python
text_numbers = {
    'one': 1,
    '1': 1,
    'two': 2,
    '2': 2,
    'three': 3,
    '3': 3,
    'four': 4,
    '4': 4,
    'five': 5,
    '5': 5,
    'six': 6,
    '6': 6,
    'seven': 7,
    '7': 7,
    'eight': 8,
    '8': 8,
    'nine': 9,
    '9': 9
}
```

```python
input_list_part2 = aoc.split_contents(aoc.read_input('Input/day_01.txt'))
```

```python
sum_of_calibration_values = 0
for test in input_list_part2:
    position_dict = dict()
    for key in text_numbers:
        position_list = list()
        for i in range(0,len(test)):
            try:
                position_list.append(test.index(key, i))
            except ValueError:
                pass
        if len(position_list) > 0:
            position_dict[key] = set(position_list)
            
    new_dict = dict()
    for key in position_dict:
        for pos in position_dict[key]:
            new_dict[pos] = text_numbers[key]
            
    number_list = sorted(new_dict.items())
    
    calibration_value = number_list[0][1] * 10 + number_list[-1][1]
    sum_of_calibration_values += calibration_value
```
